```python
from function import hankyung_data
import matplotlib.pyplot as plt
import seaborn as sns
```

## 매일 꾸준히 나오는 리포트 확인

```python
df_5 = hankyung_data('2022-05-01', '2022-05-31')
df_5.head()
```

```python
df_4 = hankyung_data('2022-04-01', '2022-04-30')
df_4.head()
```

```python
# 실제로 리포트 이름으로 갯수를 확인해 보겠습니다.
report_title_5 = df_5['REPORT_TITLE'].value_counts()
report_title_5 = report_title_5[report_title_5>3]
for i in report_title_5.index :
    print(i, report_title_5[i])
```

```python
# 실제로 리포트 이름으로 갯수를 확인해 보겠습니다.
report_title_4 = df_4['REPORT_TITLE'].value_counts()
report_title_4 = report_title_4[report_title_4>3]
for i in report_title_4.index :
    print(i, report_title_4[i])
```

```python
plt_data = report_title_4.sort_values()
plt.figure(figsize=(10,10))
plt.barh(plt_data.index, plt_data)
for i, v in enumerate(plt_data):
    plt.text(v, plt_data.index[i], v,                 # 좌표 (x축 = v, y축 = y[0]..y[1], 표시 = y[0]..y[1])
             fontsize = 9, 
             color='blue',
             horizontalalignment='left',  # horizontalalignment (left, center, right)
             verticalalignment='center')    # verticalalignment (top, center, bottom)
plt.rc('font', family='Malgun Gothic')
plt.title('4월 리포트 갯수')
plt.show()

plt_data = report_title_5.sort_values()
plt.figure(figsize=(10,10))
plt.barh(plt_data.index, plt_data)
for i, v in enumerate(plt_data):
    plt.text(v, plt_data.index[i], v,                 # 좌표 (x축 = v, y축 = y[0]..y[1], 표시 = y[0]..y[1])
             fontsize = 9, 
             color='blue',
             horizontalalignment='left',  # horizontalalignment (left, center, right)
             verticalalignment='center')    # verticalalignment (top, center, bottom)
plt.rc('font', family='Malgun Gothic')
plt.title('5월 리포트 갯수')
plt.show()
```
